## Create Dummy variables

```python
# Creates a new DF with the dummy variables

dum_data=data

for col in data.select_dtypes(['object']):
    dum_data = pd.get_dummies(dum_data, columns=[col], prefix="Dummy_"+ col)
   

```

```python
dum_data=dum_data.drop(['Dummy_Attrition_No'],axis=1)
```

```python
# Exports to CSV
dum_data.to_csv('attritionDataWithDummyVars.csv', index=False)  
```

## Normalize data + dummies

```python
ObjectData=data.select_dtypes(['object'])
NumberData=data.select_dtypes(include='number')
```

```python
from sklearn import preprocessing
NumberDataNorm = NumberData.apply(lambda iterator: ((iterator - iterator.mean())/iterator.std()).round(2))
NumberDataNorm
```

```python
NumberDataNorm.isna().sum()
```

```python
NumberData['StandardHours'].head(5)
```

```python
NumberData['EmployeeCount'].head(5)
```

```python
NumberDataNorm = NumberDataNorm.drop("StandardHours", axis=1) 
NumberDataNorm = NumberDataNorm.drop("EmployeeCount", axis=1) 
```

```python
NumberDataNorm.head()
```

```python
data_norm=NumberDataNorm.join(ObjectData)
```

```python
# Creates a new DF with the dummy variables

dum_norm_data=data_norm

for col in dum_norm_data.select_dtypes(['object']):
    dum_norm_data = pd.get_dummies(dum_norm_data, columns=[col], prefix="Dummy_"+ col)
```

```python
dum_norm_data.head()
```

```python
dum_norm_data=dum_norm_data.drop(['Dummy_Attrition_No'],axis=1)
```

```python
#shows new vars and data types
dum_norm_data.dtypes
```

```python
# Exports to CSV
dum_norm_data.to_csv('attritionDataWithDummyVarsAndNorm.csv', index=False)  
```
